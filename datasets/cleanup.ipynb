```python
import os
import hashlib
from PIL import Image
from tqdm import tqdm

## get all images in the current directory
for filename in tqdm(os.listdir("/Users/jiochoi/Documents/ComfyUI/artists")):
    if filename.endswith(".png"):
        # hash the filename
        hash = hashlib.md5(filename.replace("_1.png", "").encode()).hexdigest()

        if not os.path.exists(f"/Users/jiochoi/artists/artists/public/images/{hash}.webp"):
            # to webp
            im = Image.open("/Users/jiochoi/Documents/ComfyUI/artists/" + filename)
            im.save(f"/Users/jiochoi/artists/artists/public/images/{hash}.webp", "webp", quality=80, optimize=True)
```
